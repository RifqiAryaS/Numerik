```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameter
m = 10  # massa dalam kg
g = 10  # gravitasi m/s^2
a = 2   # percepatan m/s^2
t_total = 5  # waktu pengamatan
dt = 0.01  # delta time dalam detik (semakin kecil semakin akurat)
```

```python
# Jumlah tahapan
n_steps = int(t_total / dt)

# Arrays to store time, position, and velocity
time = np.linspace(0, t_total, n_steps)
position = np.zeros(n_steps)
velocity = np.zeros(n_steps)

# Kondisi awal
position[0] = 0
velocity[0] = 0
```

```python
# metode Euler
for i in range(1, n_steps):
    velocity[i] = velocity[i-1] + a * dt
    position[i] = position[i-1] + velocity[i-1] * dt
```

```python
# posisi terakhir
print(position[499], "m")

# kecepatan terakhir    
print(velocity[499], "m/s")
```

```python
# Plotting
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(time, position, label="Posisi (m)")
plt.xlabel("Waktu (s)")
plt.ylabel("Posisi (m)")
plt.title("Posisi terhadap Waktu")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(time, velocity, label="Kecepatan (m/s)", color='r')
plt.xlabel("Waktu (s)")
plt.ylabel("Kecepatan (m/s)")
plt.title("Kecepatan terhadap Waktu")
plt.legend()

plt.tight_layout()
plt.show()
```
